```python
df = pd.read_excel("Taxis_Data.xlsx", sheet_name="Taxis")
df['pickup'] = pd.to_datetime(df['pickup'])
df['dropoff'] = pd.to_datetime(df['dropoff'])
df['ride_duration'] = (df['dropoff'] - df['pickup']).dt.total_seconds() / 60
```

```python
# Remove invalid entries
df = df[df['distance'] > 0]
df.describe()
```

### Average fare per mile

```python
df['fare_per_mile'] = df['fare'] / df['distance']
avg_fare_per_mile = df['fare_per_mile'].mean()
avg_fare_per_mile
```

### Total revenue

```python
df['revenue'] = df['fare'] + df['tip'] + df['tolls']
total_revenue = df['revenue'].sum()
total_revenue
```

### Tip percentage

```python
df['tip_pct'] = df['tip'] / df['fare'] * 100
avg_tip_pct = df['tip_pct'].mean()
avg_tip_pct
```

### Common Pickup/Dropoff Zones
